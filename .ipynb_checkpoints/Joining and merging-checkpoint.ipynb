```python
import numpy as np
import pandas as pd
import datetime
```

# Combinig and reshaping a data

```python
customers = {'customer_id':[1234,12345,111,222,333,4444,5555,6666],
            'Name':['Sameer','Alameen','Anas','Asif','Arif','Bilal','Yousf','Basit'],
            'Address':['abc-123-2','xyz32-12','abc-123-3','xyz32-324','abc-123-453','xyz32-6343','abc-123-324','xyz32-5433']}
customers = pd.DataFrame(customers)
customers
```

```python
customers.info()
```

```python
# customers.set_index('customer_id',drop=True)
```

```python
Orders = {"customer_id": [1234,4523,4323,5324,12345,1234,12345,1234],
          'order_id': [11,22,33,44,55,66,77,88],
         'order_date':[datetime.date(2023,10,12),datetime.date(2023,10,14),datetime.date(2023,10,16),datetime.date(2023,10,18)
                      ,datetime.date(2023,10,13),datetime.date(2023,10,15),datetime.date(2023,10,17),datetime.date(2023,10,19)]}
orders = pd.DataFrame(Orders)
# orders = orders.set_index('customer_id')
orders
```

# Merge
    -Inner 
    -Outer
    -Left
    -Right

```python
merge = pd.merge(customers,orders)
# it is by default inner merge
merge
```

```python
merge1 = pd.merge(customers,orders, how='inner')
merge1 = merge1[['order_id','customer_id','Name','order_date','Address']]
merge1
```

```python
merge_data = pd.merge(customers,orders,how='outer') # outer_join >>> Union
merge_data
```

```python
merge_data = pd.merge(customers,orders,how='left')
merge_data
```

```python
merge_data = pd.merge(customers,orders, how='right')
merge_data
```

```python
customers = {'customer_id':[1234,12345,111,222,333,4444,5555,6666],
            'Name':['Sameer','Alameen','Anas','Asif','Arif','Bilal','Yousf','Basit'],
            'Address':['abc-123-2','xyz32-12','abc-123-3','xyz32-324','abc-123-453','xyz32-6343','abc-123-324','xyz32-5433']}
customers = pd.DataFrame(customers)
Orders = {"customerId": [1234,4523,4323,5324,12345,1234,12345,1234],
          'order_id': [11,22,33,44,55,66,77,88],
         'order_date':[datetime.date(2023,10,12),datetime.date(2023,10,14),datetime.date(2023,10,16),datetime.date(2023,10,18)
                      ,datetime.date(2023,10,13),datetime.date(2023,10,15),datetime.date(2023,10,17),datetime.date(2023,10,19)]}
orders = pd.DataFrame(Orders)
```

```python
merd = pd.merge(customers,orders, left_on='customer_id', right_on='customerId')
merd 
# if col_name doesnot same
```
