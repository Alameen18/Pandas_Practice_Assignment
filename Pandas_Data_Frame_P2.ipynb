# Handling Missing Values 
    Filling Missing Values
    Droping Missing values

```python
df.isnull().sum()
```

```python
df.Year.fillna(df.Year.mode().iloc[0], inplace= True)
# When we assign this column to .year, this will show nOne because inplace value always shows none and that none is assigning .Year
```

```python
df.alma_mater.fillna(df.alma_mater.mode(), inplace=True)
# In this case, this method does'nt work so we try manual one which is luckly work 
```

```python
df.alma_mater.isnull().sum()
```

```python
df.Group.fillna(df.Group.mode().mean(), inplace=True)
```

```python
df.alma_mater.fillna('US Naval Academy', inplace=True)
# This method does'nt work for categorical method so i use manual method
```

```python
df.alma_mater.isnull().sum()
```

```python
df.undergraduate_major.mode()
```

```python
df.undergraduate_major.fillna('Physics', inplace=True)
```
