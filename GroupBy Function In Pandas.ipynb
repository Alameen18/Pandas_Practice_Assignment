```python
# Load data from the websites using url for groupby Function
order_lead = pd.read_csv('https://raw.githubusercontent.com/FBosler/Medium-Data-Exploration/master/order_leads.csv',
                         parse_dates=[3] 
)
sales_team = pd.read_csv('https://raw.githubusercontent.com/FBosler/Medium-Data-Exploration/master/sales_team.csv')
    
```

```python
order_lead.head(2)
```

```python
sales_team.head(2)
```

```python
order_lead.info()
```

```python
df = pd.merge(sales_team,order_lead,on=['Company Id','Company Name'])
df.head()
```

```python
df[df.Converted == 1]
```

```python
group_by_sales = df.groupby(df['Sales Rep'].str.split(' ').str[0])
```

```python
group_by_sales['Converted'].agg('sum')
```

```python
df.groupby(pd.Grouper(key='Date', freq='QS'))['Order Value'].agg('min')
```
